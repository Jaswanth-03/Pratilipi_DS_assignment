### Sorting data based on "updated_at"

```python
user_data=user_data.sort_values(by=['updated_at'])
```

```python
user_data=user_data.drop(columns=["Unnamed: 0","updated_at"])
```

```python
user_data
```

## creating "read later" column

### considering 75% as threshold

```python
user_data["read_later"]=user_data["read_percent"].apply(lambda x: "Yes" if x>=75.0 and x<100.0 else "No")
```

```python
user_data
```

#### Splitting X and Y data

```python
X_user_data=user_data.drop(columns="read_later")
Y_user_data=user_data["read_later"]
```

```python
X_user_data
```

```python
Y_user_data
```

## Splitting dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_user_data,Y_user_data,test_size=0.25,shuffle=False,random_state=100)
```

```python
X_train
```

```python
y_train
```

### Label Encoding

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
y_train=le.fit_transform(y_train)
```

```python
y_train
```

```python
y_test=le.fit_transform(y_test)
y_test
```

## Model Building

```python
from sklearn.ensemble import RandomForestClassifier

RF_model= RandomForestClassifier(n_jobs=-2, verbose=2)
```

```python
RF_model.fit(X_train,y_train)
```

### Testing and Reports

```python
y_pred=RF_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

### Saving and Loading Model

```python
import joblib
joblib.dump(RF_model , 'model_jlib')
```

```python
loaded_model = joblib.load('model_jlib')
```
